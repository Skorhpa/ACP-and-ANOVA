```python
def drop_column(data, threshold=.5):
    return data.loc[:, data.isnull().mean() <= threshold]
```

```python
df = drop_column(dataset_OFF, 0.3)
```

```python
df.drop(df.columns[[0, 1, 2, 3, 4, 5, 6]], axis=1, inplace=True)
```

```python
df.head()
```

```python
print(df.countries_fr.nunique())
print(df.countries_tags.nunique())
print(df.countries.nunique())
```

```python
print(df.isnull().sum(axis=1))
```

```python
df = df[df.isnull().sum(axis=1) < 7]
```

```python
df
```

```python
df = df[df['product_name']!= df['product_name'].isnull()]
```

```python
df.head(5)
```

```python
#check duplicate in product_name
df.describe()
```

```python
# np.where(if, return, else)
#df['salt_100g'] = np.where(df['salt_100g'] > 100, np.nan, df['salt_100g'])

list_comp = ['fat_100g', 'saturated-fat_100g', 'carbohydrates_100g', 'sugars_100g', 'proteins_100g', 'salt_100g', 'sodium_100g']

for item in list_comp:
    df[item] = np.where(df[item] >100, np.nan, df[item])
    df[item] = np.where(df[item] <0, np.nan, df[item])
```

```python
#df['proteins_100g'] = np.where(df['proteins_100g'] >100, np.nan, df['proteins_100g'])
df.describe()
```

```python
columns_to_drop = [1, 2, 11, 12, 13]
df.drop(df.columns[columns_to_drop], axis=1, inplace=True)
```

```python
df.head()
```

```python
import pycountry

# create a new column to store the cleaned country names
df["cleaned_countries"] = df["countries"]

#iterate over the rows of the dataframe
for i, row in df.iterrows():
    try:
        country = pycountry.countries.search_fuzzy(row["countries"])[0]
        df.at[i, "cleaned_countries"] = country.name
    except:
        continue
```
