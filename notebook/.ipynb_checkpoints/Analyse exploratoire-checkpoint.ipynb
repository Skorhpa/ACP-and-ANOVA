# Analyse exploratoire – Analyse multi-variée

## ACP – Analyse multi-variée descriptive 

```python
X = df_num.values
```

```python
n_components = 10
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
# Append the principle components for each entry to the dataframe
for i in range(0, n_components):
    df['PC' + str(i + 1)] = X_pca[:, i]

display(df.head())
```

```python
print(sum(pca.explained_variance_ratio_ * 100))
```

```python
#plt.plot(np.cumsum(pca_15.explained_variance_ratio_))
exp_var= pca.explained_variance_ratio_
sing_vals=np.arange(len(exp_var)) + 1
plt.plot(sing_vals, exp_var, 'ro-', linewidth=2)
plt.xlabel('Number of components')
plt.ylabel('Explained Variance')
plt.show()
```

```python
print("Variance explained by the First principal component =", np.cumsum(pca.explained_variance_ratio_ * 100)[0])
print("Variance explained by the First 2 principal component =", np.cumsum(pca.explained_variance_ratio_ * 100)[1])
print("Variance explained by the First 3 principal component =", np.cumsum(pca.explained_variance_ratio_ * 100)[2])
print("Variance explained by the First 6 principal component =", np.cumsum(pca.explained_variance_ratio_ * 100)[5])
```

```python
# Show the points in terms of the first two PCs
g = sns.lmplot(x='PC1',
               y='PC2',
               hue='nutrition_grade_fr',data=df,
               fit_reg=False,
               scatter=True,
               height=7)

plt.show()
```

```python
# Plot a variable factor map for the first two dimensions.
(fig, ax) = plt.subplots(figsize=(8, 8))
for i in range(0, pca.components_.shape[1]):
    ax.arrow(0,
             0,  # Start the arrow at the origin
             pca.components_[0, i],  #0 for PC1
             pca.components_[1, i],  #1 for PC2
             head_width=0.05,
             head_length=0.05)

    plt.text(pca.components_[0, i] + 0.05,
             pca.components_[1, i] + 0.05,
             df_num.columns.values[i])

eigenvalues = pca.explained_variance_
eigenvectors = pca.components_.T
    
ax.grid(True, which='both')
an = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.cos(an), np.sin(an))  # Add a unit circle for scale
plt.axis('equal')
plt.xlabel('PC1 ({}%)'.format(int(round(eigenvalues[0] / sum(eigenvalues) * 100))))
plt.ylabel('PC2 ({}%)'.format(int(round(eigenvalues[1] / sum(eigenvalues) * 100))))
ax.set_title('Variable factor map')
plt.show()
```
