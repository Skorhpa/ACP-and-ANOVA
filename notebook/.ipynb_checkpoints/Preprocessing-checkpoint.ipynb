```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import time

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Importation des données

```python
%%time

dataset_OFF = pd.read_csv('../data/raw/fr.openfoodfacts.org.products.csv', sep='\t')
```

```python
#On copie le dataset d'origine pour ne pas perdre les données de base en cas d'erreur.
data = dataset_OFF.copy()
data.shape
```

# Nettoyage – Filtrage des features et produits

```python
#On affiche la somme des itérations de valeurs manquantes. C'est beaucoup, peut-être plus parlant en si l'on affiche des %.
print(data.isnull().sum())
```

```python
#On affiche les premières lignes pour voir à quoi ressemble le dataset.
data.head(4)
```

```python
#Fonction qui va retirer les colonnes ayants un nombre de valeurs manquantes fixées.
def drop_column(data, threshold=.5):
    return data.loc[:, data.isnull().mean() <= threshold]
```

```python
#On retire les colonnes qui ont plus de 50% de valeurs manquantes car jugées (arbitrairement) comme non significatives.
df = drop_column(data)
```

```python
#On réaffiche toutes les sommes de valeurs manquantes pour constater que la fonction précédente a fonctionné.
print(df.isnull().sum())
```

```python
#On enlève manuellement les colonnes jugées inutiles à l'idée finale de l'application.
#Ici, on cherche à avoir une application qui va proposer des meilleurs produits locaux (français).
#Les colonnes url, creator, created, last modified, brands_tags, ingredient_text, serving_size, states_fr et nutrition_score_uk ne sont donc pas utiles.
#De même les colonnes vagues et doublons comme coutries, coutries_tags, may_be_from_palm_oil, states, states_tags, ne sont pas utiles également.
df.drop(df.columns[[1, 2, 3, 4, 5, 6, 9, 11, 13, 14, 16, 18, 20, 21, 22, 33]], axis=1, inplace=True)
```

```python
#Il nous reste ici 18 features certaines qui vont uniquement nous servir pour du traitement de valeurs aberrantes.
print(df.isnull().sum())
```

```python
#On affiche ici le nombre de produits présents en double (longueur totale de la colonne moins le nombre de valeurs unique)
print(len(df.code)-df.code.nunique())
```

```python
#On veut maintenant filtrer les doublons en utilisant la colonne code.
df = df.drop_duplicates(subset="code", keep='first')
```

```python
#Enfin, on filtre les produits locaux uniquement donc, les produits dont la colonne countries_fr est France ou france ou fr
df = df[df['countries_fr'].str.contains('France', na=False)]
```

```python
#Pour finir, nous pouvons retirer les colonnes brands et countries car elles ne sont plus utile.
df.drop(df.columns[[2, 3, 4]], axis=1, inplace=True)
```

```python
print(df.isnull().sum())
```

# Nettoyage – valeurs aberrantes (outliers)

## Valeurs numériques

```python
#On veut nettoyer les valeurs aberrantes, nous allons d'abord nous occuper des valeurs numériques.
df.describe()
```

```python
# np.where(if, return, else)
#df['salt_100g'] = np.where(df['salt_100g'] > 100, np.nan, df['salt_100g'])
#On retire les valeurs aberrantes des colonnes des mesures sur 100g, soit toutes les valeurs négatives ou supérieures à 100. 
list_comp = ['fat_100g', 'saturated-fat_100g', 'carbohydrates_100g', 'sugars_100g', 'fiber_100g','proteins_100g', 'salt_100g', 'sodium_100g']

for item in list_comp:
    df[item] = np.where(df[item] >100, np.nan, df[item])
    df[item] = np.where(df[item] <0, np.nan, df[item])
```

```python
#Après quelques recherches, je trouve que l'aliment ayant la plus forte valeur calorique pour 100g 
#est l'huile (toute huile confondue) avec 900 calories pour 100g.
df['energy_100g'] = np.where(df['energy_100g'] > 900, np.nan, df['energy_100g'])

#De même, un nutriscore ne peut pas être négatif, quand à la valeur maximale elle ne semble pas aberrante.
df['nutrition-score-fr_100g'] = np.where(df['nutrition-score-fr_100g'] < 0, np.nan, df['nutrition-score-fr_100g'])

#Pour les ingrédients issus de l'huile de palme, après quelques recherches, il ne semble pas logique qu'il y ait plus
#d'un ingrédient qui soit issus de l'huile de palme par article, on remplace donc toutes les valeurs >1 par 1.
df['ingredients_from_palm_oil_n'] = np.where(df['ingredients_from_palm_oil_n'] >1, 1, df['ingredients_from_palm_oil_n'])
```

```python
df.describe()
```

```python
#Comme on peut voir, en numérique, il reste à traiter les colonnes additives_n, et nutriscore.

#Pour les additifs, il existe 44 types d'additifs différents qui peuvent être utilisés.
#Après consultations des valeurs supérieures à 10, il semblerait que la majorité soient des produits tels que des confiseries et des produits transformés.
plt.hist(df['additives_n'])
```

```python
#Enfin, on vérifie que les valeurs de saturated-fat_100g ne soient pas supérieures à fat_100g.
#Pareil pour le sugar_100g et carbohydrates_100g

df['saturated-fat_100g'] = np.where(df['saturated-fat_100g']>df['fat_100g'], np.nan, df['saturated-fat_100g'])

df['sugars_100g'] = np.where(df['sugars_100g']>df['carbohydrates_100g'], np.nan, df['sugars_100g'])
```

## Produits sans nom

```python
#On s'intéresse maintenant à la colonne des noms de produits, en effet si le produit ne possède pas de nom alors il n'est pas traitable.
#On retire donc toutes les lignes sans nom de produit.
#df = df[df['product_name']!= df['product_name'].isnull()]
df = df[pd.notnull(df['product_name'])]
```

```python
print(df.isnull().sum())
```

# Nettoyage – valeurs manquantes

```python
#Pour additives_n, fiber_100g et ingredients_from_palm_oil_n, on peut remplacer par 0 les valeurs manquantes car on observe
#beaucoup de valeurs nulles dans ces trois colonnes.

def replace_nan_by_value(data, column, value):
    data[column] = np.where(data[column].isnull(), value, data[column])
    return data
```

```python
df = replace_nan_by_value(df, 'additives_n', 0)
df = replace_nan_by_value(df, 'ingredients_from_palm_oil_n', 0)
df = replace_nan_by_value(df, 'fiber_100g', 0)
```

```python
#On transforme les lettres des nutrigrades en nombre pour procéder au KNN-Imputer par la suite.
abc = ['a', 'b', 'c', 'd', 'e']

for letter in range(len(abc)):
    df['nutrition_grade_fr'] = np.where(df['nutrition_grade_fr']==abc[letter], letter+1, df['nutrition_grade_fr'])

#Malgré le changement de valeur, la colonne est toujours désignée comme type objet et nous voulons la mettre en float.
df['nutrition_grade_fr'] = pd.to_numeric(df['nutrition_grade_fr'])
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import ExtraTreesRegressor
```

```python
columns = [['energy_100g', 'fat_100g', 'saturated-fat_100g'], ['carbohydrates_100g', 'sugars_100g'], ['energy_100g', 'proteins_100g'], ['salt_100g', 'sodium_100g']]

impute = IterativeImputer(estimator=ExtraTreesRegressor(n_estimators=10, random_state=0),
                          initial_strategy='most_frequent', 
                          n_nearest_features=4,
                          imputation_order='ascending')
                           #, sample_posterior=True)
    
#Fit imputer and transform
for i in columns:
    impute.fit(df[i])
    df[i] = impute.transform(df[i])
```

```python
from sklearn.impute import KNNImputer
```

```python
#On set sur un seul voisin car on souhaite un nombre entier. On va donc chercher la ligne la plus similaire en terme de 
#composition qui ne possède pas de valeur manquante et fill la valeur en fonction du résultat. Si on prend plus d'une valeur,
#Le risque est d'avoir une moyenne non entière. On pourrait faire un algorithme pour prendre l'entier le plus proche du float
#qui sortirait dans le cas d'une moyenne flottante.
imputer = KNNImputer(n_neighbors=1)
```
